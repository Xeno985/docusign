```python
import requests
import json

BASE_URL = "https://demo.docusign.net/restapi"
ACCESS_TOKEN = "your_access_token"
ACCOUNT_ID = "your_account_id"

def send_envelope(file_path, recipient_email, recipient_name):
    url = f"{BASE_URL}/v2.1/accounts/{ACCOUNT_ID}/envelopes"
    headers = {
        "Authorization": f"Bearer {ACCESS_TOKEN}",
        "Content-Type": "application/json"
    }
    with open(file_path, "rb") as file:
        base64_file = file.read().encode("base64").decode("utf-8")
    
    envelope_definition = {
        "emailSubject": "Please sign this document",
        "documents": [
            {
                "documentBase64": base64_file,
                "name": "Sample Agreement",
                "fileExtension": "pdf",
                "documentId": "1"
            }
        ],
        "recipients": {
            "signers": [
                {
                    "email": recipient_email,
                    "name": recipient_name,
                    "recipientId": "1",
                    "routingOrder": "1"
                }
            ]
        },
        "status": "sent"
    }
    response = requests.post(url, headers=headers, json=envelope_definition)
    if response.status_code == 201:
        return response.json()
    else:
        raise Exception(f"Error sending envelope: {response.status_code}")

# Example usage
response = send_envelope("contract.pdf", "recipient@example.com", "John Doe")
print(response)
```

```python
BASE_URL = "https://demo.docusign.net/restapi"
ACCESS_TOKEN = "your_access_token"
ACCOUNT_ID = "your_account_id"

def get_agreement_insights(envelope_id):
    url = f"{BASE_URL}/v2.1/accounts/{ACCOUNT_ID}/envelopes/{envelope_id}/insights"
    headers = {
        "Authorization": f"Bearer {ACCESS_TOKEN}",
        "Content-Type": "application/json"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"Error fetching insights: {response.status_code}")

# Example usage
insights = get_agreement_insights("your_envelope_id")
print(insights)
```

```python
from twilio.rest import Client

ACCOUNT_SID = "your_twilio_account_sid"
AUTH_TOKEN = "your_twilio_auth_token"

def send_sms_notification(to_number, message):
    client = Client(ACCOUNT_SID, AUTH_TOKEN)
    message = client.messages.create(
        to=to_number,
        from_="your_twilio_phone_number",
        body=message
    )
    return message.sid

# Example usage
sms_sid = send_sms_notification("+1234567890", "Contract update: Risk detected for your project.")
print(f"SMS sent with SID: {sms_sid}")
```
